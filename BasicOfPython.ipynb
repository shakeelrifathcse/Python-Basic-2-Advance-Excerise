```python
#Single Level Inheritance
class Parent:
  def __init__(self , name):
    self.name=name
    print("I am parent " , name)

  def parentMethod(self):
    print("Name is " , self.name)

class Child(Parent):
  def __init__(self,age): # consctructor with one parameter
    super().__init__("Ria") # constructor of parent class
    self.age=age
    print("I am Chils " , age)

  def childMethod(self):
    print("Age is "  , self.age)
```

```python
# In case of inheritance we cretae object of child
obj = Child(12)
# from the child object we can call parent method
obj.parentMethod()
obj.childMethod()
```

```python
class GrandParent:
  def __init__(self , name):
    self.g_name = name
class Parent(GrandParent):
  def __init__(self , p_name , g_name):
    super().__init__(g_name)
    self.p_name = p_name

class Child(Parent):
  def __init__(self ,name , p_name , g_name ):
    super().__init__(p_name , g_name)
    self.name = name

  def printName(self):
    print(self.name , self.p_name , self.g_name)
```

```python
obj = Child("Ria" , "Mahendra" ,"Icha")
obj.printName()
```

```python
# multiple inheritance
class Maternal:
  m_name = ''
  def mMethod(self , name):
    self.m_name=name
    print("Mother name is "  , self.m_name)

class Paternal:
  p_name = ''
  def pMethod(self , name):
    self.p_name=name
    print("Father name is "  , self.p_name)


class Child(Maternal , Paternal):
  def cMethod(self):
    print("Mother name is "  , self.m_name)
    print("Father name is "  , self.p_name)
```

```python
obj = Child()
obj.mMethod("Sita")
obj.pMethod("Ram")
obj.cMethod()
```

```python
# Overriding
class Rectange :
  def __init__(self , l , b):
    self.l=l
    self.b=b
  def area(self):
    print("Area of rectangle is " , self.l*self.b)

  def print_r(self):
    print("This is rectangel")

class Square(Rectange):
  def __init__(self , side):
    self.side = side

  def area(self):
    print("Area of square "  , self.side*self.side)

```

```python
s= Square(4)
s.area()

# only to call overrided method we create object of parent
r= Rectange(2,3)
r.area()
s.print_r()
```

```python
# Abstract Method : no method body
# --> Whenever a child override abstarct class it need to override or define all the abstratc method
class Shape:
  def area(self):
    pass  # blank method or method body is empty

  def perimiter(self):
    pass

class Rectangle(Shape):
  def __init__(self , l , b):
    self.l=l
    self.b=b

  #implementing the abstarct method
  def area(self):
    return self.l*self.b


class Circle(Shape):
  def __init__(self,radius):
    self.radius= radius
  def area(self):
    return 3.14*self.radius*self.radius

  def perimiter(self):
    return 2*3.14*self.radius
```

```python
rect = Rectangle(2,3)
circle=Circle(2)
print(rect.area())
print(circle.area())
print(circle.perimiter())
```

```python
# So above is not abstract method
# in python to make our method as abstract we require abstratcMethod annotation

from abc import ABC, abstractmethod
# Abstract Method : no method body
# --> Whenever a child override abstarct class it need to override or define all the abstratc method
class Shape (ABC):
  @abstractmethod
  def area(self):
    pass  # blank method or method body is empty

  @abstractmethod
  def perimiter(self):
    pass

class Rectangle(Shape):
  def __init__(self , l , b):
    self.l=l
    self.b=b

  #implementing the abstarct method
  def area(self):
    return self.l*self.b

  def perimiter(self):
    return 2*(self.l+self.b)

class Circle(Shape):
  def __init__(self,radius):
    self.radius= radius
  def area(self):
    return 3.14*self.radius*self.radius

  def perimiter(self):
    return 2*3.14*self.radius
```

```python
rect = Rectangle(2,3)
circle=Circle(2)
print(rect.area())
print(rect.perimiter())
print(circle.area())
print(circle.perimiter())
```

```python
#Custom Exception Handling
# Need to overwrit eparent class EXception
class MyException(Exception): # my class will act as exception class
  min_age=0
  def __init__(self,age):
    self.age= age
  def __str__(self): # oveeride method which print error mesage as a String
    return f"Age{self.age} should be greater than {self.min_age}"
```

```python
# Banking Exception  for withdrawl : if amount is less than available balance than raise exception  , function deposit : balance+deposit ,
# withdraw : balance - withdraw
class BankingExeption(Exception):
  def __init__(self,message):
    super().__init__(message)
```

```python
class BankAccount:
  balance = 500

  def __init__(self):
    print("Your initial balance is " , self.balance)

  def deposit(self,amount):
    if amount<=0:
      raise BankingExeption("Deposit amount more tahn 0")
    self.balance+=amount

  def withdraw(self,amount):
    if amount<=0:
      raise BankingExeption("Withdraw amount must be more than 0")
    if amount>self.balance:
      raise BankingExeption("Insufficient balance")
    self.balance-=amount
```

```python
account= BankAccount()
deposit_amount = int(input("Enter the amount"))

try:
  account.deposit(deposit_amount)
  print("Deposit successfull new balance is " , account.balance)
except BankingExeption as e:
  print(e)


try:
  withdraw_amount = int(input("Enter the amount"))
  account.withdraw(withdraw_amount)
  print("Withdraw successfull new balance is " , account.balance)
except BankingExeption as e:
  print(e)
```

```python
# Decorators : custom

def my_decorator(func):
  #create a wrapper function inside that
  def wrapper():
    print("Before function called ")
    func()
    print("After function called ")

  return wrapper

def use_wrapper ():
  print("Test the decorator")

use_wrapper1= my_decorator(use_wrapper)
use_wrapper1()
```

```python
@my_decorator
def use_wrapper():
  print("Test the decorator")

#use_wrapper2= my_decorator(use_wrapper)
#use_wrapper2()
use_wrapper()
```

```python
def sumDecorator(function):
  # always require wrapper
  def wrapper(*args):
    x,y=function(*args)
    return x+y
  return wrapper
```

```python
@sumDecorator
def sum(a,b):
  return a,b

print("Sum of integer " , sum(1,2))

print("Sum float " , sum(2.3,3.4))
```

```python
@sumDecorator
def man(a,b):
  return a,b

print(man(2,3))
```

```python
# Kwargs : keyword argument

def subDecorator(function):
  def wrapper(*args , **kwargs):
    x,y = function(*args, **kwargs)
    print("type = " , kwargs)
    return x-y
  return wrapper


@subDecorator
def man(a,b,**kwargs):
  return a,b

print(man(2,3,type="float"))
```

```python
def squareDecorator(function):
  def wrapper(*args):
    x=function(*args)
    return x*x
  return wrapper

@squareDecorator
def power(a):
  return a

print(power(2))
```

```python
# chaining of decorator
@squareDecorator
@subDecorator

def example(a,b):
  return a,b

print(example(4,8))
```

```python
# Create a decorator to convert word into upper case
def upperCase(function):
  # always require wrapper
  def wrapper(*args):
    result=function(*args)
    return result.upper()

  return wrapper


@upperCase
def get_word(name):
  return f"Hello { name }"

result = get_word("ram")
print(result)
```
