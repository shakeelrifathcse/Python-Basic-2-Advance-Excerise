```python
# Global Variables:
balance = 500
history = []

def deposit(amount):
    global balance
    balance += amount
    history.append(f'Deposited: ${amount}')
    print(f'Deposited: ${amount}, Balance: ${balance}')

def withdraw(amount):
    global balance
    if amount > balance:
        print('Insufficient Funds')
        history.append(f'Withdrawal Failed for {amount}: Insufficient Funds')
    else:
        balance -= amount
        history.append(f'Withdrawn: ${amount}')
        print(f'Withdrawn: ${amount}, Balance: ${balance}')

def check_balance():
    print(f'Current Balance: ${balance}')

def print_history():
    print('Transaction History:')
    for transaction in history:
        print(transaction)

# Using the Functions:
deposit(1000)
withdraw(500)
check_balance()
withdraw(1000)
withdraw(100)
check_balance()

print_history()
```

Exercise 13: Student Marks Analysis
Write a Python program that takes a list of students' marks (as input) and uses built-in functions
to calculate the highest, lowest, and average marks.

```python
def calculate_marks(marks):
  highest = max(marks)
  lowest = min(marks)
  average = sum(marks) / len(marks)
  return f"highest:{highest},Lowest:{lowest}, Average:{average}"

user_input = input("Enter Marks separated by spaces: ")
marks = list(map(int, user_input.split()))
calculate_marks(marks)
```

Errors:


    - Compile Time Errors
        Occurred at the time of writing a program or code.
        This can be Occurred because of
            - Syntactical Errors
                - Syntax Errors
                - Mis-spelled Keywords
            - Sementic Errors (Expressional Errors)
                Right Expression is num3 = num1 + num2
                Wrong Expression num1 + num2 = num3 (Sementic Error)

    - Logical Errors
        Occurred when we do not get the exepected output but errors are not seen.
        Example, check whether a number is even or odd.
            (num/2 == 0) ? "Even" : "Odd";
        This equation will depict that there is a logical error.

    - Run Time Errors
        Run Time Errors occurs at the time of execution may be due to wrong
         input or incorrect data format and many other reasons.
        Run Time Errors are known as exceptions.
        Execption can be handled or unhandled.
        Handled Exceptions can be caught and managed using Exception Handling.
        Exception Handling is a way of handling the run time errors or exceptions.

```python
# Basic Exception Handling with try and except

# try block - code that might raise an exception
# except block - code that handles the exception

# Example,

num1 = 100
num2 = 0
try:
    result = num1 / num2
    print("Result: ", result)
except ZeroDivisionError:
    print('{0} cannot be divided by {1}'.format(num1, num2))
    print("Error: Division by zero")



```
